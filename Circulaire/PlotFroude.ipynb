```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fr1 = np.array([7.2, 16.217, 9.9176, 4.5547, 3.0625, 4.94634, 7.0718, 9.6864])
h1 = np.array([0.002325, 0.00094616, 0.001643, 0.003987, 0.006417, 0.003589, 0.002445, 0.001666])
h2_simu = np.array([0.023, 0.0218, 0.0219, 0.0241, 0.0242, 0.0236, 0.0234, 0.022711])

Fr1_th = np.linspace(2, 20, 100)

rapport_exp = h2_simu/h1
rapport_th = 0.5*(np.sqrt(1+8*Fr1_th**2)-1)
```

```python
plt.figure(figsize=(9, 4))

plt.plot(Fr1_th, rapport_th, label=r'Théorie : $\frac{h_2}{h_1} = \frac{\sqrt{1+8 Fr_1^2}-1}{2}$', ls = '--', color='black')
plt.plot(Fr1, rapport_exp, '+', label=r'Simulations : $\frac{h_2}{h_1}$', color='red', markersize=7, marker="^")

plt.xlabel(r'Nombre de Froude aval Fr$_1$')
plt.ylabel(r'Rapport $\frac{h_2}{h_1}$')

#plt.axis("equal")
plt.title(r'Comparaison entre les simulations et la théorie pour $\frac{h_2}{h_1}$')
plt.legend()
plt.savefig('rapport_h2_h1.svg', bbox_inches='tight')
```
